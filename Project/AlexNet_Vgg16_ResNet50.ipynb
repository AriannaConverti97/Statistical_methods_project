```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from tqdm import tqdm

from tensorflow.keras.preprocessing.image import ImageDataGenerator
from model import *
from util import *
```

```python
train_dir = "data/train/"
test_dir = "data/test/"

BATCH_SIZE = 32
IMG_SIZE= 128
```

```python
trainDataGen = ImageDataGenerator(
    rotation_range=90,
    zoom_range=0.1,
    width_shift_range=0.1,
    height_shift_range=0.1,
    brightness_range = [0.5, 1.5],
    vertical_flip=True,
    fill_mode='nearest',
    rescale=1/255
)

testDataGen=ImageDataGenerator(
    rescale=1/255
)
```

### AlexNet

```python
images, labels= import_images(train_dir,224)

# Without tuning
training(
    model=alexNet(images.shape[1],images.shape[3]), 
    train_images=images, 
    train_labels=labels, 
    trainDataGen=trainDataGen, 
    testDataGen=testDataGen
)
```

```python
# With tuning
training(
    model=alexNet(images.shape[1],images.shape[3]), 
    train_images=images, 
    train_labels=labels, 
    trainDataGen=trainDataGen, 
    testDataGen=testDataGen,
    tuning=True
)
```

### VGG16

```python
images, labels= import_images(train_dir)

# Without tuning
training(
    model=vgg16(images.shape[1],images.shape[3]), 
    train_images=images, 
    train_labels=labels, 
    trainDataGen=trainDataGen, 
    testDataGen=testDataGen
)
```

```python
images, labels = import_images(train_dir)
# Without tuning
training(
    model=resNet50(images.shape[1],images.shape[3]), 
    train_images=images, 
    train_labels=labels, 
    trainDataGen=trainDataGen, 
    testDataGen=testDataGen
)
```
